```python
# needed to visualize changes in pythia
from wurlitzer import sys_pipes_forever
sys_pipes_forever()

from tqdm.notebook import tqdm

import pythia8

import numpy as np
import matplotlib.pyplot as plt
```

```python
#initializing pythia instance
pythia = pythia8.Pythia()
```

```python
#setting positron electron beam
pythia.readString("Beams:idA = 11")
pythia.readString("Beams:idB = -11")

# Function to simulate the process and retrieve the cross-section
def calculate_cross_section(energy):

    # Set up the e+e- collision process with the specified energy
    pythia.readString(f"Beams:eCM = {energy}")
    pythia.readString("WeakSingleBoson:ffbar2ffbar(s:gmZ) = on")
    pythia.readString("23:onMode = off")  # Turn off Z decays
    pythia.readString("23:onIfAny = 13 -13")  # Allow Z to decay to mu+mu-

    # Initialize Pythia
    pythia.init()

    num_events = 1000

    # Generate num_events number of events
    for iEvent in range(num_events):
        if not pythia.next():
            continue

    # Retrieve the QED total cross-section for the process
    qed_total_cross_section = pythia.info.sigmaGen()

    return qed_total_cross_section

# Energy values to consider
energies = np.logspace(1, 3, 150)
cross_sectionsArr = []

# Calculate and print cross-section for each energy
for energy in tqdm(energies):
    cross_section = calculate_cross_section(energy)
    cross_sectionsArr.append(cross_section)
    print(f"Cross-Section at {energy} GeV: {cross_section} mb")  # millibarns
```

```python
plt.plot(energies, cross_sectionsArr)
plt.yscale('log')
plt.xscale('log')
plt.xlabel('energy in (GeV)')
plt.ylabel('Cross Section (millibarns (mb))')
plt.title('Cross Section for $e^+$ $e^-$ collision ')

energy_level = 93

# Vertical line at given energy
# plt.axvline(x=energy_level, color='red', linestyle='--', label=f'Energy Level: {energy_level} GeV')
# plt.legend()
plt.grid(True)
# plt.savefig('eeminus2mumubar')
plt.show()


```
